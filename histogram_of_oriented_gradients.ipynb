# Histogram of Oriented Gradient (HOG)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

## Loading the image

```python
data_path = 'assets/hog-preprocessing.jpg'
```

```python
img = cv2.imread(data_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## Extracting the required section

```python
display(img.shape)
```

After some trial and error I could get the image section

```python
img_section = img[100:260, 205:280,:]
display(img_section.shape)
```

```python
plt.imshow(img_section)
plt.show()
```

## HOG Process

### Preprocessing

```python
image_reshaped = cv2.resize(img_section, (64, 128))
display(image_reshaped.shape)
```

```python
plt.imshow(image_reshaped)
plt.show()
```

```python
image_reshaped = np.float32(image_reshaped)/255.0
plt.imshow(image_reshaped)
plt.show()
```

### Horizontal and Vertical Gradients

```python
gx = cv2.Sobel(image_reshaped, cv2.CV_32F, 1, 0, 1)
gy = cv2.Sobel(image_reshaped, cv2.CV_32F, 0, 1, 1)
```

```python
mag, angle = cv2.cartToPolar(gx, gy, angleInDegrees=True)
```

```python
plt.imshow(gx)
plt.show()
```

```python
plt.imshow(gy)
plt.show()
```

```python
plt.imshow(mag)
plt.show()
```

```python
mag_gray = cv2.cvtColor(mag, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(mag_gray, cmap='gray')
plt.show()
```
